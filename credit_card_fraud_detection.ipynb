READ CSV FILE

```python
set=pd.read_csv('/content/creditcard.csv')
```

```python
set.head() #FIRST 5 ROWS
```

```python
set.tail()   # LAST 5 ROWS
```

```python
# DATASET INFORMATION
set.info()
```

```python
#CHECK FOR NUMBER OF MISSING VALUES IN EACH COLOUMN
set.isnull().sum()
```

CHECK THE DIFFERENT VALUES IN THE OUTPUT COLOUMN THAT IS 'CLASS

```python
set['Class'].value_counts()
```

NOW SEE THAT ONLY 1 OR 2 % DATA IS FRAUD TRANSACTION. SO MODEL CANT BE TRAINED WELL WITH THIS DATA. The data is highly unbalanced


```python
valid=set[set.Class==0]
fraud=set[set.Class==1]
```

```python
print(valid.shape)
print(fraud.shape)
```

DESCRIBE THE AMOUNT DATA

```python
valid.Amount.describe()
```

```python
fraud.Amount.describe()
```

OBSERVE HERE THAT MEAN AMOUNT FOR VALID TRANSACTION IS LESS THAN THE FRAUD ONES. THESE TYPE OF DETAILS CAN BE USED TO TRAIN THE MODEL AHEAD. NOW LETS COMPARE.

```python
#COMPARE THE VALUES FOR BOTH TYPE OF TRANSACTIONS
set.groupby('Class').mean()
```

these two rows show the means for both type of transactions. observe the difference

we have 492 fraud transaction. using random sampling we take out 492 valid transactions for balanced training

```python
 valid_sample=valid.sample(n=492)
```

```python
#concatnate two datasets. fraud and vaid_sample
newset=pd.concat([valid_sample,fraud],axis=0)
```

```python
newset.head()
```

```python
newset['Class'].value_counts()
```

```python
newset.groupby('Class').mean()
```

still the mean is very different

Now split the data into features and targets

```python
x=newset.drop(columns='Class',axis=1)
y=newset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split the data into training and test data. using traintestspli from sklearn imported at start

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training- Logistic regression for classification

```python
model=LogisticRegression()
```

```python
#train the logistic regression model
model.fit(x_train,y_train)
```

MODEL EVALUATION
 BASED ON ACCURACY SCORE

```python
#ACCURACY ON TRAINING DATA
x_train_predict=model.predict(x_train)
accuracy_train=accuracy_score(x_train_predict,y_train)
```

```python
print("Accuracy on training data: ",accuracy_train)
```

```python
#accuracy on test data
x_test_predict=model.predict(x_test)
accuracy_test=accuracy_score(x_test_predict,y_test)
```

```python
print("Accuracy on test data: ",accuracy_test)
```
